```python
import requests

def get_posts():
    url = 'https://api.met.no/weatherapi/locationforecast/2.0/compact?lat=63.430&lon=10.393'
    headers = {"User-Agent": "MyWeatherApp/1.0 (annaebr@stud.ntnu.no)"}

    response = requests.get(url, headers=headers)

    try:
        response = requests.get(url, headers=headers)

        if response.status_code == 200: #ved 200 er forespørsel velykket
            weather_data = response.json()
            return weather_data #lagrer dataen 
        
        else:
            print('Error:', response.status_code) #hvis status.code ikke er 200, da er forespøsel om api mislykket
            return None
        
    except requests.exceptions.RequestException as e:
        print('Error:', e)
        return None



get_posts()









```
